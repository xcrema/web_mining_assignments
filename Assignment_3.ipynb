```python
def closeness_centrality(G, u=None, distance=None, wf_improved=True):
    
    if G.is_directed():
        G = G.reverse()  # create a reversed graph view

    if distance is not None:
        # use Dijkstra's algorithm with specified attribute as edge weight
        path_length = functools.partial(
            nx.single_source_dijkstra_path_length, weight=distance)
    else:
        path_length = nx.single_source_shortest_path_length

    if u is None:
        nodes = G.nodes
    else:
        nodes = [u]
    closeness_centrality = {}
    for n in nodes:
        sp = path_length(G, n)
        totsp = sum(sp.values())
        len_G = len(G)
        _closeness_centrality = 0.0
        if totsp > 0.0 and len_G > 1:
            _closeness_centrality = (len(sp) - 1.0) / totsp
            # normalize to number of nodes-1 in connected part
            if wf_improved:
                s = (len(sp) - 1.0) / (len_G - 1)
                _closeness_centrality *= s
        closeness_centrality[n] = _closeness_centrality
    if u is not None:
        return closeness_centrality[u]
    else:
        return closeness_centrality
```

```python
def proximity_prestige(G, u=None, distance=None, wf_improved=False):
    if distance is not None:
        # use Dijkstra's algorithm with specified attribute as edge weight
        path_length = functools.partial(
            nx.single_source_dijkstra_path_length, weight=distance)
    else:
        path_length = nx.single_source_shortest_path_length

    if u is None:
        nodes = G.nodes
    else:
        nodes = [u]
    proximity_prestige = {}
    for n in nodes:
        sp = path_length(G, n)
        c=0
        for k in nodes:
            if(nx.has_path(G,k,n)):
                c=c+1
        totsp = sum(sp.values())
        len_G = len(G)
        _proximity_prestige = 0.0
        if totsp > 0.0 and len_G > 1:
            _proximity_prestige = -1*(c/len(G)-1)/(totsp/c) #(len(sp) - 1.0) / totsp
            # normalize to number of nodes-1 in connected part
        proximity_prestige[n] = _proximity_prestige
    if u is not None:
        return proximity_prestige[u]
    else:
        return proximity_prestige
```

```python
n=len(adj)
sum=0

for i in range(n):
    temp=0
    for j in range(n):
        temp += adj[i][j]
    if temp==sum:
        for j in range(n):
            adj[i][j]=1/n
print(adj)
            
        
```

```python
E=np.ones((n,n))
sto=((1-d)*E/n)+d*np.transpose(adj)
print(sto)
```

```python
P=[]
P.append(np.ones((n,1))/n)
k=0
e=0.05
c=1
while c:
    P.append(np.dot(sto, P[k]))
    a=((P[k+1]-P[k]))<e
    stop = a.all()
    if stop:
        c=0
             
    k=k+1
```

```python
count = 0
for i in range(len(P)):
    count+=1
    print("\n\n\n\n")
    print(f'Iteration {i}')
    print(P[i])
    print("\n\n\n\n")
```
