```python
file = open("demo.txt", "a+")
```

```python
ch = 'y'
while(ch == 'y'):
    temp = input("Enter Line: ")
    file.write(temp+"\n")
    ch = input("Wanna input more lines?: ")
```

```python
file.seek(0, 0)
```

```python
groups = {}
row = 0
col = 0
for line in file:
    temp = line.split()
    row+=1
    for word1 in temp:
        for word2 in temp:   
            if(word1 == word2):
                if word1 not in groups.keys():
                    groups[word1] = list()
                groups[word1].append((row, col))
            col+=len(word2)+1
        col=0
            
(file.readline()).split()
```

```python
groups
```

```python
arrr = {"Words": list(groups.keys()), "Positions": list(groups.values())}
```
